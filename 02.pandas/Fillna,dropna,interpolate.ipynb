```python
new_df = df.fillna(0)      #用0替换NAN
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
new_df = df.fillna(method="ffill")     #copying previous values
new_df
```

```python
new_df = df.fillna(method="bfill")    #copying below values
new_df
```

```python
new_df = df.fillna(method="bfill",axis='columns')    #copying right values
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)   
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
